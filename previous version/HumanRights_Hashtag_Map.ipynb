```python
import matplotlib.pyplot as plt
from plots import generate_list
from pyvis.network import Network as net
```

```python
import networkx as nx
```

```python
human_rights_list = generate_list("hashtags", "humanrights", "2021-06-10", "2021-06-20")
human_rights_list
```

```python
def get_weights(hashtag):
    return round(human_rights_list.count(hashtag) / len(human_rights_list), 3)
```

```python
##### Example 1 #####
```

```python
hashtag_graph = nx.Graph()
hashtag_graph.add_node("humanrights", pos=(30, 4))
i = 2
j = 0
for hashtag in human_rights_list[-10:]:
    if not hashtag_graph.has_node(hashtag):
        hashtag_graph.add_node(hashtag, pos=(j, i))
        hashtag_graph.add_edge("humanrights", hashtag, weight=get_weights(hashtag))
        if i == 0:
            i = 2
        else:
            i -= 1
        j += 10
```

```python
pos=nx.get_node_attributes(hashtag_graph,'pos')
labels = nx.get_edge_attributes(hashtag_graph,'weight')
nx.draw(hashtag_graph, pos, with_labels=True)
nx.draw_networkx_edge_labels(hashtag_graph, pos, edge_labels=labels)
nt = net.Network(notebook=True)
nt.from_nx(hashtag_graph)
# nt.enable_physics(True)
nt.show('tree.html')
```

```python
##### Example 2 #####
```

```python
all_hashtag_graph = nx.Graph()
all_hashtag_graph.add_node("humanrights")
for hashtag in human_rights_list:
    if not all_hashtag_graph.has_node(hashtag):
        all_hashtag_graph.add_node(hashtag)
        all_hashtag_graph.add_edge("humanrights", hashtag, weight=get_weights(hashtag))
```

```python
nx.draw(all_hashtag_graph, with_labels=True)
```
